```python
import torch
import numpy as np
from config import device,TRAIN_CONFIG
from data.data_loader import get_data_loaders
from training.optimizer_comparison import OptimizerComparator
from utils.visualization import show_samples, plot_optimizer_comparison

```

```python
def main():
    # 设置随机种子
    torch.manual_seed(TRAIN_CONFIG['random_seed'])
    np.random.seed(TRAIN_CONFIG['random_seed'])
    
    print("Fashion-MNIST 分类任务 - 优化器比较研究")
    print(f"使用设备: {device}")
    print("=" * 60)
    
    # 加载数据
    print("加载数据...")
    train_loader, test_loader = get_data_loaders()
    
    # 显示样本
    print("显示训练样本...")
    show_samples(train_loader)
    
    # 比较优化器
    print("开始比较不同优化器...")
    comparator = OptimizerComparator(device)
    results = comparator.compare_optimizers(
    train_loader, test_loader,
    tune_lr=True,       # 开启学习率搜索
    lr_tune_epochs=5    # 每个候选 lr 跑 5 个 epoch
)
    
    # 显示结果摘要
    comparator.print_summary()
    
    # 可视化结果
    print("\n生成可视化图表...")
    plot_optimizer_comparison(results)

if __name__ == "__main__":
    main()
```

```python
import matplotlib.pyplot as plt
from torchvision import datasets, transforms
from data.data_loader import get_class_names

# 1. 加载 Fashion-MNIST 测试集（只做 ToTensor）
transform = transforms.ToTensor()
test_dataset = datasets.FashionMNIST(
    root="./data",
    train=False,
    download=True,
    transform=transform,
)

class_names = get_class_names()          # ['T-shirt/top', 'Trouser', ...]
num_classes = len(class_names)

# 2. 为每个类别找一张样本图像
images_per_class = [None] * num_classes

for img, label in test_dataset:
    if images_per_class[label] is None:
        images_per_class[label] = (img, label)
    # 如果 10 个类别都已经找到样本就停止
    if all(x is not None for x in images_per_class):
        break

# 3. 画出 10 个类别的示例图
plt.figure(figsize=(12, 6))
for idx, (img, label) in enumerate(images_per_class):
    plt.subplot(2, 5, idx + 1)
    plt.imshow(img.squeeze(), cmap="gray")
    plt.title(class_names[label])
    plt.axis("off")

plt.tight_layout()
plt.show()
```
